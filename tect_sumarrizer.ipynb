```python
##prompt template
from langchain.chains import LLMChain
from langchain import PromptTemplate

generictemplaye = """
Write a summary  of the following speech:
Speech: {speech}

translate the summary in language: {language}
"""

prompt = PromptTemplate(
    input_variables = ['speech','language'],
    template = generictemplaye
)
prompt 
```

```python
complete_prompt=prompt.format(speech = speech,language = 'Hinglish')
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain = LLMChain(llm = llm, prompt = prompt)
summary = llm_chain.run({'speech':speech,'language':'Hinglish'})
summary
```

# Stuff Document Chain Text Summmarization


```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("apjspeech.pdf")
docs = loader.load_and_split()
docs
```

```python
template = """
Write a consise and short summary of the  following speech,
Speech :{text}
"""

prompt = PromptTemplate(input_variables = ['text'],template = template)
```

```python
from langchain.chains.summarize import load_summarize_chain
```

```python
chain = load_summarize_chain(llm,chain_type="stuff",prompt = prompt,verbose=True)
output =chain.run(docs)
output
```

# Map reduce to summarize Large Documents

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
docs
```

```python
final_docs = RecursiveCharacterTextSplitter(chunk_size = 2000,chunk_overlap = 200).split_documents(docs)
final_docs
```

```python
len(final_docs)
```

```python
chunk_prompt  = """ 
plaese summarixe the speech.
Speech : {text}
"""
map_prompt_template = PromptTemplate(input_variables = ['text'],template = chunk_prompt)
```

```python
final_prompt = """
Provide the final summary of the entire speech with the important speech 
Add a MMotivational Title,Start the precise summary with an introduction and end it with a conclusion.
Speech: {text}
"""
final_prompt_template=PromptTemplate(input_variables = ['text'],template = final_prompt)
final_prompt_template
```

```python
prompt = PromptTemplate(input_variables = ['text'],template = template)
summary_chain = load_summarize_chain(llm,
                                     chain_type="map_reduce",
                                     map_prompt = map_prompt_template,
                                     combine_prompt = final_prompt_template,
                                     verbose=True)
output =summary_chain.run(final_docs)
output
```

# Refine Chain for summarization 


```python
chain = load_summarize_chain(llm=llm,
                             chain_type="refine",
                             verbose=True)
output =chain.run(final_docs)
output
```
